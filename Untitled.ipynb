```python
import re
import pandas as pd

# baca file SQL
with open("stock_transactions.sql", "r") as file:
    sql = file.read()

# ambil isi dalam tanda kurung (VALUES)
values = re.findall(r'\((.*?)\)', sql)

# ubah ke list
data = [v.split(",") for v in values]

# bersihkan data
cleaned = []
for row in data:
    cleaned.append([item.strip().replace("'", "") for item in row])

# buat dataframe
df = pd.DataFrame(cleaned)

df.head()
```

```python
df.columns = ["id", "product_id", "transaction_date", "quantity_change", "transaction_type"]
```

```python
df.to_csv("stock_transactions.csv", index=False)
```
